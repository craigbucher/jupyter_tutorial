```python
import numpy as np
import pandas as pd
# import csv file
my_df = pd.read_csv('dog_data_short.csv')
my_df
```

```python
# Add column from list
gender = ['Male', 'Female', 'Male', 'Male', 'Female'] # must match number of rows in existing datafile
my_df['Gender'] = gender
my_df
```

```python
# Add default values
my_df['Alive/Dead'] = [True]*len(my_df) # So that it adds to every existing row
my_df
```

```python
# Add NaN value (say, if you don't know the value)
my_df['Show Dog'] = [np.nan]*len(my_df) # np.nan - creates NaN value (just using 'NaN' doesn't work)
my_df
```

```python
# Add column with .insert() - allows position
my_df.insert(1, 'Adopted', [True]*len(my_df), True) # (position, column header, value, allow duplicates def=True)
my_df
```

```python
# Add column with .assign() - ** creates new df **
my_df2 = my_df.assign(Horse = [False]*len(my_df))  # Are these dogs horses?
my_df2
```
