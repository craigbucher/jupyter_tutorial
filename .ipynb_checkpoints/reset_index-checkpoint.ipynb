```python
import numpy as np
import pandas as pd
# import csv file
my_df = pd.read_csv('dog_data_short.csv')
my_df
```

```python
# Create new column
my_df['Frame Header'] = ['Dog1', 'Dog2', 'Dog3', 'Dog4', 'Dog5']
my_df
```

```python
# Set index - turn 'Frame Header' column items into row indices
my_df.set_index('Frame Header', inplace=True) # use 'inplace=True' to make it permanent
my_df
```

```python
# Reset index - turn back to sequential numbers
my_df.reset_index(inplace=True)
my_df
```

```python
# Drop 'Frame Header' column
my_df.drop('Frame Header', axis=1, inplace=True) # 'axis=1' = columns
my_df
```
